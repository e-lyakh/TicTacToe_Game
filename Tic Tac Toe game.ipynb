```python
from IPython.display import clear_output

def chose_players():
    
    players = {}
    
    while True:        
        
        user_input = input("Choose a sign for Player #1 (X or O): ")
          
        if user_input.upper() == 'X':
            players = {'Player 1':'X','Player 2':'O', 'Current':'Player 1'}
            return players
        elif user_input.upper() == 'O':
            players = {'Player 1':'O','Player 2':'X', 'Current':'Player 1'}
            return players
        else:            
            clear_output()
            print("Please, input a correct sign (X or O)")
            continue
```

```python
def display_board(moves):
    
    print('     |     |     ' + '  ||  ' + '     |     |     ')
    print('  1  |  2  |  3  ' + '  ||  ' + '  ' + moves[0] + '  |  ' + moves[1] + '  |  ' + moves[2] + '  ')
    print('     |     |     ' + '  ||  ' + '     |     |     ')
    print('-----------------' + '  ||  ' + '-----------------')
    print('     |     |     ' + '  ||  ' + '     |     |     ')
    print('  4  |  5  |  6  ' + '  ||  ' + '  ' + moves[3] + '  |  ' + moves[4] + '  |  ' + moves[5] + '  ')
    print('     |     |     ' + '  ||  ' + '     |     |     ')
    print('-----------------' + '  ||  ' + '-----------------')
    print('     |     |     ' + '  ||  ' + '     |     |     ')
    print('  7  |  8  |  9  ' + '  ||  ' + '  ' + moves[6] + '  |  ' + moves[7] + '  |  ' + moves[8] + '  ')
    print('     |     |     ' + '  ||  ' + '     |     |     ')
```

```python
def change_player(players):
    
    if players['Current'] == 'Player 1':
        players['Current'] = 'Player 2'
    else:
        players['Current'] = 'Player 1'
        
    return players
```

```python
def accept_player_move(players, moves):    
    
    while True:        
        
        display_board(moves)
        
        user_num_choice = input(players['Current'] + ": enter your move (1-9): ")        
        
        if user_num_choice.isdigit() == False:
            clear_output()
            print("Please, enter a correct input (a number 1-9)!")            
            continue
            
        move_id = int(user_num_choice) - 1
        
        if move_id not in range(0,9):
            clear_output()
            print("Please, enter a correct number (1-9)!")
            continue
            
        if moves[move_id] != ' ':
            clear_output()
            print("The place you chose (" + user_num_choice  + ") is busy. Choose another one")
            continue
            
        moves[move_id] = players[players['Current']]
        
        clear_output()      
        
        return moves    
```

```python
def check_winner(moves):    
    
    def is_winning_row(row):
        return ' ' not in row and len(set(row)) == 1    
    
    if is_winning_row(moves[0:3]):
        return moves[0]
    
    if is_winning_row(moves[3:6]):
        return moves[3]
    
    if is_winning_row(moves[6:9]):
        return moves[6]
    
    if is_winning_row([moves[0],moves[3],moves[6]]):
        return moves[0]
    
    if is_winning_row([moves[1],moves[4],moves[7]]):
        return moves[1]
    
    if is_winning_row([moves[2],moves[5],moves[8]]):
        return moves[2]
    
    if is_winning_row([moves[0],moves[4],moves[8]]):
        return moves[0]
    
    if is_winning_row([moves[2],moves[4],moves[6]]):
        return moves[2]
    
    if ' ' not in moves:
        return 'D'
    
    return 'N'
```

```python
def show_winner(check_winner_result, players):    
    
    if check_winner_result == players['Player 1']:        
        print('Player 1 won. Congratulations!')
        return
    
    if check_winner_result == players['Player 2']:        
        print('Player 2 won. Congratulations!')
        return
    
    if check_winner_result == 'D':
        print("It's a draw.")
        return
    
    print("Something wrong")
```

```python
def should_continue_game():
    
    while True:
        
        user_input = input("Do you want to continue the game ? (Y or N): ")
        
        if user_input.upper() != 'Y' and user_input.upper() != 'N':
            clear_output()
            print("Please, enter a correct answer (Y or N): ")
            continue
            
        clear_output()
        
        return user_input.upper() == 'Y'
```

```python
def play_game():
    
    while True:        
        
        players = chose_players()
        
        moves = [' '] * 9
        
        while True:
            
            moves = accept_player_move(players, moves)            
            
            winner = check_winner(moves)
            
            if winner != 'N':
                break
            
            change_player(players)            
        
        show_winner(winner, players)
        
        if should_continue_game():
            continue
        else:
            return        
```

```python
play_game()
```
